```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('Dataset for Data Analytics.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.fillna('')
print("Filled NULL values with empty string")
```

```python
dataset["CustomerID"].duplicated().sum()
```

```python
dataset = dataset.drop_duplicates(subset="CustomerID")
print("Dropped duplicate Customer IDs")
```

```python
dataset["CustomerID"].duplicated().sum()
```

```python
dataset["Date"].isnull().sum()
```

```python
dataset["Date"] = pd.to_datetime(dataset["Date"], errors="coerce")
dataset["Date"].isnull().sum()
print("No invalid format for Date found")
```

```python
dataset.isnull().sum()
print("NULL values have been replace by empty string")
```

```python
dataset.duplicated().sum()
print("Duplicate values have been emitted")
```
